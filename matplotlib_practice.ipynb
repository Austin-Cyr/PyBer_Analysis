```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i*5 for i in y_axis], color='skyblue', label='Boston', edgecolors = 'black', linewidth = 3)
ax.legend()
ax.set_title("Fare by Month")
ax.set_xlim(10,40)
ax.invert_yaxis()
```
